## Exploration of nodes time delay 
Better understanding of parameters signalV and Dmat

```python
from neurolib.utils.loadData import Dataset

ds = Dataset("hcp")
```

```python
delay = 250
signalV = 10
Cmat = np.array([[0, 1],[0, 0]])
Dmat = delay*np.ones((2,2))
wc = WCModel(Cmat = Cmat, Dmat = Dmat)
wc.params['exc_ext'] = 0.65
wc.params['signalV'] = signalV
wc.params['duration'] = 6 * 1000 
wc.params['sigma_ou'] = 0.14
wc.params['K_gl'] = 3.15
wc.run(chunkwise=True)
rhos = []
for d in range(10,300,10):
    rho = np.corrcoef([wc.exc[0,d:],wc.exc[1,:-d]])
    rhos.append(rho[0,1])
plt.plot(range(10,300,10), rhos);
plt.title(f"Correlation first nodes excitation with shifted second, \n for delay : {delay} and signalV: {signalV}");
```
